# Membuat Centeroid

```python
import random as rn

k = 2
clusters = {}
# for i in range(k):
#     clusters[i] = []
# print("cluster : ",clusters)

centroids = []

# RANDOM GENERATE CENTROID
for y in range(k):
    print("\nCentroid ke-", y)
    random_pick_centroid = rn.randint(0, 30)
    print("Ambil Baris Random : ", random_pick_centroid)
    centroids.append(list_data[random_pick_centroid])
    print("Nilai Array[pm10][pm25] Centroid : ", centroids[y], "\n----------------")

print("\n========================\n")
#AMBIL INDEX 25 dan 20
# y=0
# print("\nCentroid ke-", y)
# print("Angka Index : ", 25)
# centroids.append(list_data[25])
# print("Array Centroid : ", centroids[y], "\n----------------")  
# y+=1
# print("\nCentroid ke-", y)
# print("Angka Index : ", 20)
# centroids.append(list_data[20])
# print("Array Centroid : ", centroids[y], "\n----------------")  
# print(centroids)

def recalculate_clusters(length_ispu, centroids, k):
    # Initiate clusters kosong
    clusters = {}
    # Set k
    for i in range(k):
        clusters[i] = []
    for data in range(length_ispu):
        euc_dist = []
#         print("\n====================================================================================\nbaris ke-",data,"\n------------")
        for j in range(k):
            cpm10=centroids[j][0]
            cpm25=centroids[j][1]

#             print(">> Array centroid ke-",j," : ",centroids[j])
#             print(">> SQRT( (",cpm10,"-",pm10[data],")^2 + (",cpm25,"-",pm25[data],")^2 )")

            euc_dist.append(np.sqrt(pow((cpm10-pm10[data]),2)+pow((cpm25-pm25[data]),2)))  

#             print("----------------------------------------------------------------------------------")
            print("==> Jarak Antar Centeroid-",j," baris ke-",data,": ",euc_dist[j])
            print("\n")
        clusters[euc_dist.index(min(euc_dist))].append(data)
#         print("========>> Kelompok clusters : ", clusters,"\n====================================================================================\n")
    
    recalculate_centroids(centroids, clusters, k)
    return clusters

def recalculate_centroids(centroids, clusters, k):
    pm10_average = []
    pm25_average = []
    for i in range(k):
        print("-> centroids ke -",i,centroids[i])
#         print("-> clusters kelompok-",i," = ",clusters[i])
        pm10_list = [] 
        pm25_list = [] 
        length_clusters = len(clusters[i])
#         print("Pm10  Pm25\n---- ----- ")
        for j in range(length_clusters):
            index_cluster = clusters
            pm10_anggota = pm10[index_cluster[i][j]]
            pm10_clusters = pm10_anggota 
            pm10_list.append(pm10_clusters)

            pm25_anggota = pm25[index_cluster[i][j]]
            pm25_clusters = pm25_anggota 
            pm25_list.append(pm25_clusters)
#             print("(",j,")", pm10_clusters,"-",pm25_clusters)
#             print(pm10_list,"-",pm25_list)
        pm10_average.append(np.average(pm10_list, axis=0))
        pm25_average.append(np.average(pm25_list, axis=0))
#         print("----------------\n\n",pm10_average[i], pm25_average[i])  
        centroids[i] = [pm10_average[i], pm25_average[i]]
#         centroids[i] = np.average(pm10_list, axis=0)
    print("\n-> clusters kelompok-",0," = ",clusters[0])
    print("-> clusters kelompok-",1," = ",clusters[1])
    print("\n==> Nilai Centeroid Baru didapat = ",centroids)
    return centroids

length_ispu = len(list_data)

for iterate in range(9):
    recalculate_clusters(length_ispu, centroids, k)
#     print(clusters)
    print('\n====================================================================================================\n')
```
